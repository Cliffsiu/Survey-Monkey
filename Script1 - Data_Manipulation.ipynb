Make a copy. Sometimes the file is quite large. If we make a mistake, then we can make a copy of the data instead of loading it again.

```python
df_modified = df.copy()
df_modified
```

```python
df_modified.columns
```

```python
list_to_drop = ['Start Date', 'End Date', 'Email Address',
       'First Name', 'Last Name', 'Custom Data 1']
df_modified = df_modified.drop(columns=list_to_drop)
df_modified
```

```python
id_var = list(df_modified.columns[:8])
value_var = list(df_modified.columns[8:])
```

```python
df_melted = df_modified.melt(id_vars=id_var, value_vars=value_var, var_name="Question+Subquestion", value_name="Answer")
df_melted
```

```python
question_import = pd.read_excel(pwd + "\\Data - Survey Monkey Output edit.xlsx", sheet_name = "Question")
question_import
```

```python
question = question_import.copy()
question = question.drop(columns=["Raw Subquestion", "Raw Question", "Subquestion"])
question = question.dropna()
question
```

```python
df_merged = pd.merge(left=df_melted, right=question, how="left", left_on="Question+Subquestion", right_on="Question+Subquestion")
df_merged
```

### Aggregation(group)

```python
respondents = df_merged[df_merged["Answer"].notna()]
respondents = respondents.groupby("Question")["Respondent ID"].nunique().reset_index()
respondents.rename(columns={"Respondent ID": "Respondents"}, inplace=True)
respondents
```

```python
df_merged_v2 = pd.merge(left=df_merged, right=respondents, how="left", left_on="Question", right_on="Question")
df_merged_v2
```

```python
same_answer = df_merged            #[df_merged["Answer"].notna()]
same_answer = same_answer.groupby(["Question+Subquestion", "Answer"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID": "Same Answer"}, inplace=True)
same_answer
```

```python
df_merged_v3 = pd.merge(left=df_merged_v2, right=same_answer, how="left", left_on=["Question+Subquestion", "Answer"], right_on=["Question+Subquestion", "Answer"])
df_merged_v3["Same Answer"].fillna(0, inplace=True)
df_merged_v3
```

```python
output = df_merged_v3.copy()
output.rename(columns={'Identify which division you work in.-Response':"Division Primary",
       'Identify which division you work in.-Other (please specify)':"Division Secondary",
       'Which of the following best describes your position level?-Response':"Position",
       'Which generation are you apart of?-Response':"Generation",
       'Please select the gender in which you identify.-Response':"Gender",
       'Which duration range best aligns with your tenure at your company?-Response':"Tenure",
       'Which of the following best describes your employment type?-Response':"Employment Type"}, inplace=True)
output
```

```python
output.to_excel(pwd + "\\Final_Output.xlsx", index=False)
```
